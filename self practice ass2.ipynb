```python
roll_no=[]
hindi_marks=[]
science_marks=[]
maths_marks=[]
english_marks=[]
history_marks=[]
goegraphy_marks=[]
result=dict()
divisions=dict()


fp1=open("results.csv","r")
data=fp1.readline()

while(True):   
   
    data=fp1.readline()
    if not data:
        break;
    #print(data)
    data=data.replace("\n","")
    temp=data.split(",")
    roll_no.append(temp[0])
    hindi_marks.append(temp[1])
    science_marks.append(temp[3])
    maths_marks.append(temp[4])
    english_marks.append(temp[2])
    history_marks.append(temp[5])
    goegraphy_marks.append(temp[6])
    result.update({temp[0]:temp[7]})
    divisions.update({temp[7]:temp[8]})
  
fp1.close()

maths_marks=tuple(maths_marks)
print(type(maths_marks))
```

```python
print("\nhindi marks\n",hindi_marks,end="")
print("\n\nscience marks\n",science_marks,end="")
print("\n\nmaths marks\n",maths_marks,end="")
print("\n\nenglish markss\n",english_marks,end="")
print("\n\nhistory markss\n",history_marks,end="")
print("\n\nresults\n",result,end="")
print("\n\ndivision wise \n",divisions,end="")



```

```python
topper = {}
for item in result:
   
   if item in topper:
      
      topper[item] += 1
   else:
      
     topper[item] = 1

print(topper)
marklist = sorted(topper.items(), key=lambda x:x[1],reverse=True)
sortdict = dict(marklist)
print(sortdict)
print("The most marks of students",list(sortdict.keys())[0]," are ",list(sortdict.values())[0],"this")
```
